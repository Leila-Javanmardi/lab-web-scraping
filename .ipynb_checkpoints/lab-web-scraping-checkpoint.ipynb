````python
# scrape_imdb that takes four parameters: title_type,user_rating, start_date, and end_date.
def scrape_imdb(title_type,user_rating, start_date, end_date):
    # Lists to store extracted data
    Movie_Nr=[]
    Title=[]
    Year=[]
    Rating=[]
    Runtime=[]
    Genre=[]
    Description=[]
    Director=[]
    Stars=[]
    Votes=[]
    Gross=[]
    
    url1 = f"https://www.imdb.com/list/ls002206094/?sort=list_order,asc&st_dt=&mode=detail&page=1&title_type={title_type}&user_rating={user_rating}%2C&release_date={start_date}%2C{end_date}&ref_=ttls_vm_dtl"
    
            #https://www.imdb.com/list/ls002206094/?sort=list_order,asc&st_dt=&mode=detail&page=1&title_type=movie&user_rating=7.5%2C&release_date=1990%2C1992&ref_=ttls_vm_dtl
    
    
    response2 = requests.get(url1)
    response2
    soup2 = BeautifulSoup(response2.content, "html.parser")

    # Find all components
    components = soup2.find_all('div', class_='lister-item-content')

    for component in components:
        # Extracting info form component
        

        number = component.find('span', class_="lister-item-index unbold text-primary").text.replace('.','').strip()
        Movie_Nr.append(number)

        title = component.find('a').text.strip()
        Title.append(title)

        year = component.find('span', class_="lister-item-year text-muted unbold").text.strip('()')
        Year.append(year)

        rating = component.find('span', class_="ipl-rating-star__rating").text.strip()
        Rating.append(rating)
        
        runtime = component.find('span', class_="runtime").text.strip()
        Runtime.append(runtime)

        genre = component.find('span', class_='genre').text.strip()
        Genre.append(genre)

        description = component.find_all('p')[1].text.strip()
        Description.append(description)

        director = component.find_all('a')[12].text.replace("\n","").strip()
        #director = component.find_all('p')[2].text.strip().split()[1:5]
        Director.append(director)
        
        for i in range(13,17):
            star = component.find_all('a')[i].text.replace("\n","").strip()
            Stars.append(star)
        
        joined_stars = ', '.join(Stars)
        
        #none of the methods bellow give me the name properly -> ask Ignacio
        #star = component.find_all('a')[13].text.replace("\n","").strip()
        #star = component.find_all('p',text='Stars:').text.strip()
        #star = component.find_all('p')[2].text.strip().split()[5:]
        #star = component.find('span', text='Stars:').text.strip()
        #Stars.append(star)
    

        votes = component.find_all('span')[59].text.strip()                                                             
        Votes.append(votes)
        

        gross = component.find_all('span')[62].text.replace("$","").replace("M","").strip()
        Gross.append(gross)
        

    # Create DataFrame
    df = pd.DataFrame({
        
        'Movie_Nr':Movie_Nr,
        'Title':Title,
        'Year':Year,
        'Rating':Rating,
        'Runtime':Runtime,
        'Genre':Genre,
        'Description':Description,
        'Director':Director,
        'Stars':joined_stars,
        'Vote':Votes,
        'Gross':Gross
        
        })
    
    return df
````
